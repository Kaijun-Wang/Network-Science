## 常见的度分布：泊松分布与幂律分布

## 泊松分布：以ER随机网络为例

```python
# 导入库
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# 创建一个ER随机网络
n = 10000
p = 0.001
ER = nx.erdos_renyi_graph(n,p)
```

```python
# 获取平均度
d = dict(nx.degree(ER))
print("平均度为：",sum(d.values())/len(ER.nodes))
```

```python
# 获取所有可能的度值对应的概率
x = list(range(max(d.values())+1))
y  = [i / n for i in nx.degree_histogram(ER)]
```

```python
# 绘制度分布
plt.plot(x,y,"ro-")
plt.xlabel("$k$")
plt.ylabel("$p_k$")
```

## 幂律分布：以BA无标度网络为例

```python
m  = 3
BA = nx.barabasi_albert_graph(n,m)
```

```python
# 获取平均度
d = dict(nx.degree(BA))
print("平均度为：",sum(d.values())/len(BA.nodes))
```

```python
# 获取所有可能的度值对应的概率
x = list(range(max(d.values()) + 1))
y  = [i / n for i in nx.degree_histogram(BA)]
```

```python
# 绘制度分布
plt.plot(x,y,"ro-")
plt.xlabel("$k$")
plt.ylabel("$p_k$")
```

```python
# 在双对数坐标轴下显示
plt.plot(x,y,"ro-")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("$k$")
plt.ylabel("$p_k$")
```

```python
# 在双对数坐标轴下要把横坐标和纵坐标的0值排除掉（log0=∞）
new_x = []
new_y = []
for i in range(len(x)):
    if y[i]!=0:
        new_x.append(x[i])
        new_y.append(y[i])
        
plt.plot(new_x,new_y,"ro-")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("$k$")
plt.ylabel("$p_k$")
```
